## **degree 3 rise**

4.   then use fit_transform
5.   again fit into a linear regression

```python
from sklearn.model_selection import train_test_split  #for splitting 80 and 20
from sklearn.linear_model import LinearRegression
from sklearn import metrics #accuracy
```

```python
x=da1.drop('Y house price of unit area',axis=1)
x
```

```python
y=da1['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False) #include_bias is used to avoid noise
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape

```

```python
x_train,x_test,y_train,y_test,=train_test_split(poly_feat,y,test_size=0.3,random_state=101)
```

```python
polymodel=LinearRegression()
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y Test': y_test,'Y_pred':y_pred,'Residuals':(y_test-y_pred)}).head()
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)    #loss function
poly_RMSE=np.sqrt(poly_MSE)
```
