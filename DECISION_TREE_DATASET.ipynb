## **factorize - label encoding**
label encoding - a,b,c == 0,1,2


```python
da['class'],class_names=pd.factorize(da['class'])
print(class_names)
```

```python
da_backup=da.copy()
da_backup.head()
```

```python
da['buying'],_=pd.factorize(da['buying'])
da['maint'],_=pd.factorize(da['maint'])
da['doors'],_=pd.factorize(da['doors'])
da['persons'],_=pd.factorize(da['persons'])
da['lug_boot'],_=pd.factorize(da['lug_boot'])
da['safety'],_=pd.factorize(da['safety'])
da.head()
```

```python
x = da[['buying','maint','doors','persons','lug_boot','safety']]
y=da[['class']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.6,random_state=0)  # usually split value will be taken in random but after using random_state=0 it will be taken in continuous
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_test_pred=dtree.predict(x_test)
accuracy=metrics.accuracy_score(y_test,y_test_pred)
print('Test accuracy:{:.2f}'.format(accuracy))
```
